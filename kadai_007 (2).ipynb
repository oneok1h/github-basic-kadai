```python
# ライブラリーをインポートする
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
import numpy as np

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
print(df)

# カテゴリー列の出現頻度をカウントする
category_count = df['カテゴリー'].value_counts()
print(category_count)

# 棒グラフを作成する
plt.figure(figsize=(10, 6))
plt.bar(category_count.index, category_count.values)
plt.xlabel('カテゴリー')
plt.ylabel('出現頻度')
plt.title('カテゴリー列の要素の出現頻度')
plt.xticks(rotation=45)
plt.show()

# グループ化して集計し、記述統計を得る
quantity_count = df.groupby('商品番号')['カテゴリー'].count()
quantity_desc = df.groupby('商品番号')['カテゴリー'].describe()

# quantiry_countとquantity_descを表示
print(quantity_count)
print(quantity_desc)
```
